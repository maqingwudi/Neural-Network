```python
import numpy as np
import matplotlib.pyplot as plt
import random
class DNHHNetwork:
    def __init__(self, patterns):
        self.num_neurons = patterns.shape[1]
        self.weights = np.zeros((self.num_neurons, self.num_neurons))  # Initialize weight matrix to zeros



    
    def train(self, patterns):
        """
        Train the Hopfield network. The input is a set of patterns, each being a vector representing the state of neurons.
        Each pattern will be used as a learning sample for the network.
        """
        for id,pattern in enumerate(patterns):
            #pattern = np.array(pattern)
            if (id == 0):
                self.weights = np.outer(pattern, pattern)
            else:
                self.weights += np.outer(pattern, pattern)  # Accumulate the outer product

        # Dynamic normalization: prevent the weight matrix from becoming too large
        np.fill_diagonal(self.weights, 0)  # Set diagonal elements to 0
        #self.weights /= np.max(np.abs(self.weights))  # Dynamically normalize the weights by the maximum absolute value

    def run(self, input_pattern, max_iterations=100):
        """
        Run the Hopfield network. Given an initial input pattern, the network will iterate and update its state until convergence.
        """
        pattern = np.array(input_pattern)
        for _ in range(max_iterations):
            for i in range(self.num_neurons):
                net_input = np.dot(self.weights[i], pattern)
                pattern[i] = 1 if net_input > 0 else -1  # Apply threshold function
            # Check for convergence
            if np.array_equal(pattern, np.sign(np.dot(self.weights, pattern))):
                break
        return pattern

```

```python
def add_noise_to_pattern(pattern, noise_ratio=0.1):

    pattern = np.array(pattern) 
    noise_count = int(len(pattern) * noise_ratio)
    noise_positions = random.sample(range(len(pattern)), noise_count)
    noisy_pattern = pattern.copy()
    for pos in noise_positions:
        noisy_pattern[pos] = 1 - noisy_pattern[pos]
    return noisy_pattern
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hopfield
def compute_hetero_associative_weights(input_patterns, target_patterns):
    """
    Hopfield

    
    - input_patterns: 
    - target_patterns: 
    
    
    - W
    """
    num_patterns = input_patterns[0].shape[0]
    num_units = target_patterns[0].shape[0]

    # 
    W = np.zeros((num_patterns, num_units))

    # 
    for i,t in zip(input_patterns,target_patterns):
        #print(i)
        W+=np.outer(i,t)
        #print(np.outer(i,t))
    
    
    return W

# Hopfield
def hetero_associative_update(W, input_pattern,max_iterations=100):
    """
    

    
    - W: 
    - input_pattern: 

    
    - 
    """
    num_units = len(W.T)
    output_pattern = np.zeros(len(W.T))
    # 
    for _ in range(max_iterations):
        for i in range(num_units):
            sum_input = np.dot(W.T[i], input_pattern)  # i
            #print("sum",sum_input)
            # 
            output_pattern[i] = 1 if sum_input > 0 else -1
        if np.array_equal(output_pattern, np.sign(np.dot(W.T, input_pattern))):
            break
    return output_pattern

def add_noise(vector, noise_level):
    vector = np.asarray(vector)  # Ensure vector is a NumPy array for efficient processing
    n_flip = int(len(vector) * noise_level)
    
    if n_flip == 0:
        return vector
    
    flip_indices = np.random.choice(len(vector), n_flip, replace=False)
    vector[flip_indices] *= -1  # In-place flip of the selected indices
    
    return vector

def is_equal(train_result,test_result):
    return np.equal(train_result,test_result)
# 
# -1  +1 
letters_2 = {
    'G': G,  
    'N': N,
    'I': I,
    'Q': Q,
    'O': O,
    'U': U,
    'H': H,
    'Z': Z
}
data_x_1 = [np.array(letters_2[letter])*2-1 for letter in letters_2.keys()]
data_y_1 = [
    np.array([-1, -1, -1]),  # G
    np.array([-1, -1, 1]),   # N
    np.array([-1, 1, -1]),   # I
    np.array([-1, 1, 1]),    # Q
    np.array([1, -1, -1]),   # O
    np.array([1, -1, 1]),    # U
    np.array([1, 1, -1]),    # H
    np.array([1, 1, 1])      # Z
]

# data_x=np.array(data_x)*2-1
# data_y=np.array(data_y)
# Hopfield
W = compute_hetero_associative_weights(data_x_1, data_y_1) #35x3
#print(W)
# 
# print("Input Patterns:\n", data_x_1)
# print("Target Patterns:\n", data_y_1)
data_x_noise = add_noise(data_x_1,0.1)
# 
for id,i in enumerate(data_x_1):
    output_pattern_test = hetero_associative_update(W, i,100)
    print(list(letters_2.keys())[id])
    print("Test Input Pattern_x:", i)
    print("Output Pattern after Hetero-associative Recall:", output_pattern_test)
    print("Is equal:",is_equal(output_pattern_test,data_y_1[id]))
    # plt.subplot(1, 2, 1)
    # plt.imshow(np.array(input_pattern_test).reshape(7, 5), cmap='binary')
    # plt.title("Training Pattern Noise_ratio=0.2")

print("Output test")
for id,i in enumerate(data_y_1):
    output_pattern_test = hetero_associative_update(W.T, i,100)
    print(list(letters_2.keys())[id])
    print("Test Input Pattern_y:", i)
    print("Output Pattern after Hetero-associative Recall:", output_pattern_test)
    print("Is equal:",is_equal(output_pattern_test,data_x_1[id]))
    # plt.subplot(1, 2, 2)
    #plt.imshow(np.array(output_pattern_test).reshape(7, 5), cmap='binary')
#print("Test Input Pattern:", input_pattern_test)

## 
for id,i in enumerate(data_x_noise):
    output_pattern_test = hetero_associative_update(W, i,100)
    print(list(letters_2.keys())[id])
    print("Test Input Pattern_x_noise:", i)
    print("Output Pattern after Hetero-associative Recall:", output_pattern_test)
    print("Is equal_noise:",is_equal(output_pattern_test,data_y_1[id]))


```
